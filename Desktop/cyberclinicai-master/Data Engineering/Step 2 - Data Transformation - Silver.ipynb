```
#creating variables and assigning the values
spark.conf.set('DI.TW',2)
spark.conf.set('DI.ALL_S',1)
spark.conf.set('DI.ALL_ES',2)
#logic to find the appointments with rE returning status
re_df = spark.sql("select * from (select  t.index_id, t.appointment_date, t.patient_id, row_number() over (partition by t.patient_id order by t.appointment_date asc) rn, months_between(t.appointment_date, lag(appointment_date) over(partition by patient_id order by appointment_date asc)) month_diff from updates t ) where month_diff >= '${DI.TW}'");
#creating temp view with rE data
re_df.createOrReplaceTempView('updates_re')
```

```
#Enriching returning status, previous_session and succssive_session columns from the data
f_df = spark.sql("""select index_id, patient_id,
            practitioner_id,
            appointment_date,
            social_wellbeing,
            work_wellbeing,
            patient_gender,
            goals_performance,
            appointment_id,
            audio_record_file_url,
            empathy_performance,
            personal_wellbeing,
            family_wellbeing,
            confidence_performance,
            patient_age,
            case
            when de_date > appointment_date and count(case when de_date > appointment_date then 1 else null end) over (partition by patient_id) = ${DI.ALL_S} then 'D'
            when de_date <= appointment_date then 'rE'
            when app_count = ${DI.ALL_S} and months_diff < ${DI.TW}  then 'X'
            when app_count = ${DI.ALL_S} and months_diff >= ${DI.TW} then 'D'
            when app_count >= ${DI.ALL_ES} then 'E' end returning_status,
            previous_session,
            successive_session,
            fees, 
            practitioner_age, 
            secondary_problem_area, 
            practitioner_experience, 
            patient_type, 
            primary_problem_area,
            practitioner_gender,
            session_number,
            medicare_rebate from
(select max(re.appointment_date) over (partition by f.patient_id) de_date, re.patient_id re_pat_id, 
        max(session_number) over (partition by f.patient_id) app_count,
        months_between(current_date(), f.appointment_date) months_diff,
        count(*) over(partition by f.patient_id order by f.appointment_date ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) - 1 previous_session ,
        count(*) over(partition by f.patient_id order by f.appointment_date ROWS BETWEEN CURRENT ROW and UNBOUNDED following) - 1 successive_session,
        f.* from (select distinct * from updates) f left outer join updates_re re on f.index_id = re.index_id) t 
        order by t.patient_id asc, t.appointment_date asc""")
```

```
#creating temp view with enriched data
f_df.createOrReplaceTempView('updates_final')
```
