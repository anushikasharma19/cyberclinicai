```
df = spark.sql("SELECT * FROM cyb.silver_data").drop("model_trained")
display(df)
```

```
unique_paient = spark.sql("SELECT * FROM cyb.silver_data").select("patient_id").distinct().collect()
unique_paient
```

```
max_df = spark.sql("SELECT patient_id, MAX(session_number) FROM cyb.silver_data GROUP BY patient_id ORDER BY patient_id").toPandas()
max_df
```

```
pandas_df = df.toPandas()
pandas_df
```

```
lst_of_patient = []
for i in unique_paient:
  get_patient = pandas_df[pandas_df['patient_id']==i[0]]
  get_max_session = max_df[max_df['patient_id']==i[0]]['max(session_number)'].values[0]
  if len(get_patient) == get_max_session:
    lst_of_patient.append(get_patient)
```

```
import pandas as pd
main_df = pd.concat(lst_of_patient)
main_df
```

```
df = spark.createDataFrame(main_df)
df.createOrReplaceTempView("temporary_table")
display(df)
```

```
%sql
SELECT * FROM temporary_table WHERE social_wellbeing=0 AND work_wellbeing=0 AND goals_performance= 0 AND empathy_performance= 0 AND personal_wellbeing=0 AND family_wellbeing=0 AND confidence_performance=0 ORDER BY patient_id, session_number
```

```
df_without_0 = spark.sql("SELECT * FROM temporary_table WHERE index_id NOT IN ( SELECT index_id FROM temporary_table WHERE social_wellbeing=0 AND work_wellbeing=0 AND goals_performance= 0 AND empathy_performance= 0 AND personal_wellbeing=0 AND family_wellbeing=0 AND confidence_performance=0)")
```

```
df_without_0.createOrReplaceTempView("temp_without_0")
```

```
average_df = spark.sql("SELECT patient_id, AVG(social_wellbeing), AVG(work_wellbeing), AVG(goals_performance), AVG(empathy_performance), AVG(personal_wellbeing), AVG(family_wellbeing),AVG(confidence_performance) FROM temp_without_0 GROUP BY patient_id")
display(average_df)
```

```
average_df = average_df.toPandas()
```

```
pandas_df = df.toPandas()
pandas_df
```

```
for i in range(len(pandas_df)):
  if (pandas_df.loc[i]['social_wellbeing'] == 0 and pandas_df.loc[i]['work_wellbeing'] == 0 and pandas_df.loc[i]['goals_performance'] == 0 and pandas_df.loc[i]['empathy_performance'] == 0 and pandas_df.loc[i]['personal_wellbeing'] == 0 and pandas_df.loc[i]['family_wellbeing'] == 0 and pandas_df.loc[i]['confidence_performance'] == 0):
    highest_session = max(pandas_df[pandas_df['patient_id']== pandas_df.loc[i]['patient_id']]['session_number'].tolist())
    if pandas_df.loc[i]['session_number'] != highest_session:
      pandas_df.loc[i,'social_wellbeing'] = float(average_df[average_df['patient_id']==pandas_df.loc[i]['patient_id']]['avg(social_wellbeing)'].tolist()[0])
      pandas_df.loc[i,'work_wellbeing'] = float(average_df[average_df['patient_id']==pandas_df.loc[i]['patient_id']]['avg(work_wellbeing)'].tolist()[0])
      pandas_df.loc[i,'goals_performance'] = float(average_df[average_df['patient_id']==pandas_df.loc[i]['patient_id']]['avg(goals_performance)'].tolist()[0])
      pandas_df.loc[i,'empathy_performance'] = float(average_df[average_df['patient_id']==pandas_df.loc[i]['patient_id']]['avg(empathy_performance)'].tolist()[0])
      pandas_df.loc[i,'personal_wellbeing'] = float(average_df[average_df['patient_id']==pandas_df.loc[i]['patient_id']]['avg(personal_wellbeing)'].tolist()[0])
      pandas_df.loc[i,'family_wellbeing'] = float(average_df[average_df['patient_id']==pandas_df.loc[i]['patient_id']]['avg(family_wellbeing)'].tolist()[0])
      pandas_df.loc[i,'confidence_performance'] = float(average_df[average_df['patient_id']==pandas_df.loc[i]['patient_id']]['avg(confidence_performance)'].tolist()[0])
```

```
temp = spark.createDataFrame(pandas_df)
temp.createOrReplaceTempView("testing")
```

```
%sql
SELECT * FROM testing WHERE social_wellbeing=0 AND work_wellbeing=0 AND goals_performance= 0 AND empathy_performance= 0 AND personal_wellbeing=0 AND family_wellbeing=0 AND confidence_performance=0 ORDER BY patient_id, session_number
```

```
%sql
SELECT * FROM temporary_table WHERE patient_id IN (SELECT patient_id FROM testing WHERE social_wellbeing=0 AND work_wellbeing=0 AND goals_performance= 0 AND empathy_performance= 0 AND personal_wellbeing=0 AND family_wellbeing=0 AND confidence_performance=0) ORDER BY patient_id,session_number
```

```
temp.write.format('delta').mode('overwrite').option("overwriteSchema", "true").partitionBy('patient_id').saveAsTable('cyb.analysing_data_temporary')
```
